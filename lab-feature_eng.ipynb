**Check for missing values**

```python
spaceship.isnull().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship = spaceship.dropna()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship["Cabin"].nunique()
```

```python
spaceship["Cabin"].unique()
```

```python
spaceship["Cabin"] = spaceship["Cabin"].str[0]
```

```python
spaceship["Cabin"].unique()
```

- Drop PassengerId and Name

```python
spaceship = spaceship.drop(columns=["PassengerId", "Name"])
```

- For non-numerical columns, do dummies.

```python
spaceship.info()
```

```python
spaceship
```

```python
spaceship_dummies = pd.get_dummies(spaceship, columns=["HomePlanet", "CryoSleep", "Cabin", "Destination"], drop_first = False)

spaceship_dummies
```

```python
spaceship_dummies.info()
```

```python
columns_to_transform = ["Transported", "HomePlanet_Earth", "HomePlanet_Mars", "HomePlanet_Europa", "CryoSleep_False", "CryoSleep_True", "Cabin_A", "Cabin_B", "Cabin_C", "Cabin_D", "Cabin_E", "Cabin_F", "Cabin_G", "Cabin_T", "Destination_55 Cancri e", "Destination_PSO J318.5-22", "Destination_TRAPPIST-1e", "VIP"]
```

```python
spaceship_dummies[columns_to_transform] = spaceship_dummies[columns_to_transform].applymap(lambda x: 1 if x else 0)

spaceship_dummies
```

**Perform Train Test Split**

```python
features = spaceship_dummies.drop(columns = ["Age"])

target = spaceship_dummies["Age"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size = 0.20, random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
normalizer = MinMaxScaler()
```

```python
normalizer.fit(X_train)
```

```python
X_train_norm = normalizer.transform(X_train)

X_test_norm = normalizer.transform(X_test)
```

```python
X_train_norm = pd.DataFrame(X_train_norm, columns = X_train.columns)
X_train_norm.head()
```

```python
X_test_norm = pd.DataFrame(X_test_norm, columns = X_test.columns)
X_test_norm.head()
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors = 2)
```

```python
knn.fit(X_train_norm, y_train)
```

- Evaluate your model's performance. Comment it

```python
knn.score(X_test_norm, y_test)
```
